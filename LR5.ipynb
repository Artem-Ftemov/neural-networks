Імпортуємо залежності

```python
import tensorflow as tf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import image_dataset_from_directory
```

Створюємо функцію переводу тензора до нового виду

```python
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label
```

Вказуємо папку зі зображеннями

```python
base_dir = '/content/img/img'
```

Генеруємо датасет з папки із зображеннями та використовуємо функцію "process" до кожного елементу датасету

```python
train_datagen = image_dataset_from_directory(base_dir, image_size=(100,100), seed = 1, batch_size= 128, label_mode='categorical', color_mode='grayscale')
train_datagen = train_datagen.map(process)
```

Створюємо модель

```python
model = tf.keras.models.Sequential([
    layers.Flatten(),
    layers.Dense(1024, activation='relu'),
    layers.Dense(512, activation='LeakyReLU'),
    layers.Dense(64, activation='LeakyReLU'),
    layers.Dense(14, activation='softmax')
])
```

Ініціалізуємо змінну із функцією оптимізації та компілюємо модель

```python
opt = keras.optimizers.Adam(learning_rate=0.0005)

model.compile(
    loss='categorical_crossentropy',
    optimizer=opt,
    metrics=['accuracy']
)
```

```python
model.summary()
```

Тренуємо модель

```python
history = model.fit(train_datagen,
                    epochs=50,
                    use_multiprocessing = True,
                    shuffle = True
                   )
```
