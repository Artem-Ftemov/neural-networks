Імпортуємо залежності

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Activation, Dropout, Flatten, Dense, Conv2D, MaxPooling2D
from keras.utils import image_dataset_from_directory
```

Створюємо функцію переводу тензора до нового виду

```python
def process(image,label):
    image = tf.cast(image/255. ,tf.float32)
    return image,label
```

Вказуємо шлях до зображень на яких буде відбуватись тренування та тестування

```python
train_dir = '/content/dogs_vs_cats/train'
test_dir = '/content/dogs_vs_cats/test'
```

Генеруємо датасет з папки із зображеннями та використовуємо функцію "process" до кожного елементу датасету

```python
train_dataset = image_dataset_from_directory(train_dir, image_size=(100,100),seed = 1, batch_size= 128, label_mode='binary', color_mode='rgb')
train_dataset = train_dataset.map(process)

test_dataset = image_dataset_from_directory(test_dir, image_size=(100,100),seed = 1,batch_size= 128,label_mode='binary', color_mode='rgb')
test_dataset = test_dataset.map(process)
```

Створюємо модель із декількох шарів

```python
model = tf.keras.models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape = (100, 100, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'), 
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Conv2D(32, (3, 3), activation='relu'), 
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Flatten(),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

Ініціалізуємо змінну із функцією оптимізації та компілюємо модель

```python
optimizer = keras.optimizers.Adam(learning_rate=0.0005)

model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
```

Тренеруємо модель

```python
model.fit(train_dataset,epochs=20,use_multiprocessing = True,shuffle = True,validation_data=test_dataset)
```
